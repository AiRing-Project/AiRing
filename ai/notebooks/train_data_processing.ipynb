```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline
from kss import split_sentences
from tqdm import tqdm

# KOTE 43개 감정 → 13개 감정(형용사형+중립) 매핑 #"외로운' 없음.
kote_to_13 = kote_to_13 = {
    '불평/불만': '분노한',
    '환영/호의': '감사한',
    '감동/감탄': '감사한',
    '지긋지긋': '분노한',
    '고마움': '감사한',
    '슬픔': '슬픈',
    '화남/분노': '분노한',
    '존경': '감사한',
    '기대감': '흥분한',
    '우쭐댐/무시함': '자신하는',
    '안타까움/실망': '슬픈',
    '비장함': '자신하는',
    '의심/불신': '불안한',
    '뿌듯함': '만족스러운',
    '편안/쾌적': '편안한',
    '신기함/관심': '흥분한',
    '아껴주는': '감사한',
    '부끄러움': '당황한',
    '공포/무서움': '불안한',
    '절망': '우울한',
    '한심함': '우울한',
    '역겨움/징그러움': '분노한',
    '짜증': '분노한',
    '어이없음': '분노한',
    '없음': '그저 그런',
    '패배/자기혐오': '우울한',
    '귀찮음': '스트레스 받는',
    '힘듦/지침': '스트레스 받는',
    '즐거움/신남': '기쁜',
    '깨달음': '만족스러운',
    '죄책감': '슬픈',
    '증오/혐오': '분노한',
    '흐뭇함(귀여움/예쁨)': '기쁜',
    '당황/난처': '당황한',
    '경악': '당황한',
    '부담/안_내킴': '스트레스 받는',
    '서러움': '슬픈',
    '재미없음': '우울한',
    '불쌍함/연민': '슬픈',
    '놀람': '흥분한',
    '행복': '기쁜',
    '불안/걱정': '불안한',
    '기쁨': '기쁜',
    '안심/신뢰': '편안한'
    # '외로운'에 직접적으로 해당하는 라벨은 KOTE 원본에 없음
}


selected_emotions = [
    "기쁜", "감사한", "편안한", "만족스러운", "흥분한", "자신하는",
    "분노한", "슬픈", "우울한", "불안한", "스트레스 받는", "당황한", "그저 그런"
]

def chunk_text_with_kss(text, tokenizer, max_tokens=512):
    sentences = split_sentences(str(text))
    chunks = []
    current_chunk = []
    current_length = 0
    for sentence in sentences:
        encoded = tokenizer.encode(sentence, add_special_tokens=True)
        tokenized_len = len(encoded)
        if tokenized_len > max_tokens:
            # 단어 단위로 강제 분할
            words = sentence.split()
            chunk = []
            chunk_len = 0
            for word in words:
                word_encoded = tokenizer.encode(word, add_special_tokens=False)
                word_len = len(word_encoded)
                if chunk_len + word_len > max_tokens:
                    chunks.append(' '.join(chunk))
                    chunk = []
                    chunk_len = 0
                chunk.append(word)
                chunk_len += word_len
            if chunk:
                chunks.append(' '.join(chunk))
            current_length = 0
            current_chunk = []
            continue
        if current_length + tokenized_len <= max_tokens:
            current_chunk.append(sentence)
            current_length += tokenized_len
        else:
            chunks.append(' '.join(current_chunk))
            current_chunk = [sentence]
            current_length = tokenized_len
    if current_chunk:
        chunks.append(' '.join(current_chunk))
    return chunks

def kote_predict(text, pipe, kote_to_13, threshold=0.4, top_n=3, final_threshold=0.6):
    tokenizer = pipe.tokenizer
    chunks = chunk_text_with_kss(text, tokenizer)
    chunk_results = []
    all_emotions = []
    for chunk in chunks:
        outputs = pipe(chunk)
        mapped_scores = []
        for output_list in outputs:
            for out in output_list:
                mapped = kote_to_13.get(out['label'])
                if mapped in selected_emotions and out['score'] >= threshold:
                    mapped_scores.append((mapped, out['score']))
        mapped_scores = sorted(mapped_scores, key=lambda x: x[1], reverse=True)
        seen = set()
        top_emotions = []
        for emotion, score in mapped_scores:
            if emotion not in seen:
                top_emotions.append((emotion, score))
                seen.add(emotion)
            if len(top_emotions) >= top_n:
                break
        chunk_results.append(dict(top_emotions))
        all_emotions.extend([e for e, _ in top_emotions])
    total_emotions = {}
    for cr in chunk_results:
        for k, v in cr.items():
            if k not in total_emotions or v > total_emotions[k]:
                total_emotions[k] = v
    if not total_emotions or all(v < threshold for v in total_emotions.values()):
        emotions_prob = {'그저 그런': 1.0}
        exist_emotions = ['그저 그런']
    else:
        emotions_prob = total_emotions
        # 확률 내림차순 정렬 후 상위 3개만
        sorted_emotions = sorted(
            [(emotion, prob) for emotion, prob in emotions_prob.items() if prob >= final_threshold],
            key=lambda x: x[1], reverse=True
        )
        exist_emotions = [emotion for emotion, prob in sorted_emotions[:3]]
        if not exist_emotions:
            exist_emotions = ['그저 그런']
    return emotions_prob, exist_emotions



# ------------------- CSV 파일 처리 -------------------

input_csv = "/content/drive/MyDrive/diary_sentiment_analysis/train_df.csv"
output_csv = "/content/drive/MyDrive/diary_sentiment_analysis/train_df_emotions_1.csv"

df = pd.read_csv(input_csv)

model_name = "searle-j/kote_for_easygoing_people"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
pipe = TextClassificationPipeline(
    model=model,
    tokenizer=tokenizer,
    function_to_apply='sigmoid',
    return_all_scores=True,
    device=0,  # GPU 사용
    truncation=True,  # 추가
    max_length=512    # 추가
)

emotions_prob_list = []
exist_emotions_list = []

for text in tqdm(df['text'], desc="감정 분석 중"):
    try:
        emotions_prob, exist_emotions = kote_predict(
            text, pipe, kote_to_13, threshold=0.4, top_n=3, final_threshold=0.6
        )
    except Exception as e:
        print(f"Error: {str(e)}")
        emotions_prob = {'error': 1.0}
        exist_emotions = ['error']
    emotions_prob_list.append(str(emotions_prob))
    exist_emotions_list.append(str(exist_emotions))

df['emotions_prob'] = emotions_prob_list
df['exist_emotions'] = exist_emotions_list

df.to_csv(output_csv, index=False)
print(f"분석 결과가 저장되었습니다: {output_csv}")
```

#전체 코드 단계별 로직
##1. 데이터 및 모델 준비
CSV 파일에서 일기 텍스트(text 컬럼) 불러오기

KOTE 멀티라벨 감정 분류 모델 및 토크나이저 로드

KOTE 원본 43개 감정 → 13개 감정(형용사형+중립)으로 매핑 테이블 정의

##2. 텍스트 전처리
KSS 문장 분리기로 텍스트를 자연스러운 문장 단위로 분리

각 문장의 토큰 길이를 계산 (tokenizer.encode)

512 토큰 이하(특수 토큰 포함)로 여러 문장을 묶어 chunk 생성

한 문장이 너무 길면 단어 단위로 강제 분할하여 chunk 생성

##3. 감정 예측 및 후처리
각 chunk별로 감정 예측 (멀티라벨, sigmoid 확률)

모델이 반환한 원본 감정명(label)을 kote_to_13 매핑을 통해 13개 감정군으로 변환

threshold(예: 0.4) 이상 감정만 "존재"로 인정, 상위 top_n(예: 3개)만 추출

chunk별로 감정별 최대 확률 집계

##4. 최종 감정 결정
모든 감정 확률이 threshold 미만이면 '그저 그런'만 남김

아니면, 존재하는 감정 중 final_threshold(예: 0.6) 이상 감정만 최종 감정으로 인정 (내림차순 정렬 후 상위 3개)

최종 감정이 없다면 '그저 그런'만 남김

##5. 결과 저장
감정별 확률 딕셔너리(emotions_prob)와 최종 감정 리스트(exist_emotions)를 새 컬럼에 저장

결과를 새로운 CSV 파일로 저장

```python
# Jaccard 유사도(다중 감정 일치도) 계산

from typing import List

def jaccard_similarity(list1: List[str], list2: List[str]) -> float:
    set1, set2 = set(list1), set(list2)
    if not set1 and not set2:
        return 1.0
    return len(set1 & set2) / len(set1 | set2)

# 각 행마다 Jaccard 유사도 계산
jaccard_scores = [
    jaccard_similarity(a, b)
    for a, b in zip(pre_top500['exist_emotions'], gemini_df['exist_emotions'])
]

# 평균 유사도 계산
mean_jaccard = sum(jaccard_scores) / len(jaccard_scores)
print(f"평균 Jaccard 유사도: {mean_jaccard:.3f}")
```

```python
# 감정별 분포 비교

import ast
from collections import Counter
import pandas as pd

# 문자열을 리스트로 변환
pre_top500['exist_emotions'] = pre_top500['exist_emotions'].apply(ast.literal_eval)
gemini_df['exist_emotions'] = gemini_df['exist_emotions'].apply(ast.literal_eval)

# 평탄화 및 집계
all_emotions_pre_top500 = [emotion for emotions in pre_top500['exist_emotions'] for emotion in emotions]
emotion_counts_pre_top500 = Counter(all_emotions_pre_top500)
result_pre_top500 = pd.DataFrame(emotion_counts_pre_top500.items(), columns=['emotion', 'count']).sort_values('count', ascending=False)

all_emotions_gemini_df = [emotion for emotions in gemini_df['exist_emotions'] for emotion in emotions]
emotion_counts_gemini_df = Counter(all_emotions_gemini_df)
result_gemini_df = pd.DataFrame(emotion_counts_gemini_df.items(), columns=['emotion', 'count']).sort_values('count', ascending=False)

print("pre_top500 감정별 분포")
print(result_pre_top500)

print("\ngemini_df 감정별 분포")
print(result_gemini_df)

```

### 평균 Jaccard 유사도: 0.526

해석:

다중 감정 분석에서 이 정도 유사도는, 감정 분류 체계가 동일하더라도 모델별 감정 해석 방식이 상당히 다름을 보여줍니다.


### 감정별 분포 차이

- 사전학습 모델이 더 많이 잡은 감정:
편안한, 감사한, 스트레스 받는, 흥분한, 기쁜, 당황한  

- LLM 모델이 더 많이 잡은 감정:
그저 그런, 만족스러운, 불안한, 우울한, 슬픈, 자신하는, 분노한  

- 특이점:  

        '그저 그런': LLM 모델은 160건, 사전학습 모델은 7건 → LLM 모델이 중립적 감정을 훨씬 더 많이 매핑

        '만족스러운': LLM 모델이 더 많이 잡음

        '편안한', '감사한', '스트레스 받는', '흥분한': 사전학습 모델이 상대적으로 더 많이 매핑


#### 해석 및 인사이트
- 모델별 감정 해석 경향 차이

  LLM 모델은 중립적이거나 복합적인 상황에 '그저 그런', '만족스러운', '불안한', '우울한', '슬픈' 등 감정을 더 적극적으로 부여

  사전학습 모델은 '기쁜', '감사한', '편안한', '스트레스 받는', '흥분한' 등 긍정적·활동적 감정을 더 자주 매핑

- 중립/복합 감정 처리

  LLM 모델이 '그저 그런' 등 중립 감정을 더 많이 감지하는 것은, 감정의 미묘함이나 애매함을 더 포착하려는 경향으로 볼 수 있음

- 감정 분포의 다양성

  LLM 모델은 감정 분포가 더 넓고, 사전학습 모델은 특정 감정에 집중되는 경향

```python
# 감정별 대표 사례 분석
# "같은 일기에서 Hugging Face는 '기쁜'으로, Gemini는 '만족스러운'으로 감정을 매핑(분류)한 사례"

for idx, (hf_emotions, gm_emotions) in enumerate(zip(pre_top500['exist_emotions'], gemini_df['exist_emotions'])):
    if '기쁜' in hf_emotions and '만족스러운' in gm_emotions and '기쁜' not in gm_emotions:
        print(f"Index: {idx}")
        print("일기 내용:", pre_top500.iloc[idx]['text'])
        print("Hugging Face 감정:", hf_emotions)
        print("Gemini 감정:", gm_emotions)
        print("-" * 40)
        # 필요시 break로 일부만 출력
```

```python
# 슬픈과 우울한 감정이 매핑된 일기 샘플 추출
for idx, (hf_emotions, gm_emotions) in enumerate(zip(pre_top500['exist_emotions'], gemini_df['exist_emotions'])):
    # 한 모델은 '슬픈', 다른 모델은 '우울한'만 있는 경우
    # Hugging Face와 Gemini 모두 '슬픈', '우울한'을 포함하거나, 한 모델은 '슬픈', 다른 모델은 '우울한'을 포함한 경우

    if ('슬픈' in hf_emotions and '우울한' in gm_emotions) or ('우울한' in hf_emotions and '슬픈' in gm_emotions):
        print(f"Index: {idx}")
        print("일기 내용:", pre_top500.iloc[idx]['text'])
        print("Hugging Face 감정:", hf_emotions)
        print("Gemini 감정:", gm_emotions)
        print("-" * 40)
```

```python
# Gemini에서 '그저 그런'이 매핑된 사례 추출
for idx, (hf_emotions, gm_emotions) in enumerate(zip(pre_top500['exist_emotions'], gemini_df['exist_emotions'])):
    if '그저 그런' in gm_emotions:
        print(f"Index: {idx}")
        print("일기 내용:", pre_top500.iloc[idx]['text'])
        print("Hugging Face 감정:", hf_emotions)
        print("Gemini 감정:", gm_emotions)
        print("-" * 40)
```

#### 1. 모델별 감정 해석 경향  
Hugging Face 모델  

즉각적이고 명확한 긍정(‘기쁜’, ‘감사한’, ‘흥분한’ 등)이나 부정(‘슬픈’, ‘스트레스 받는’ 등) 감정을 주로 매핑.

감정의 복합성이나 중립성(‘그저 그런’)보다는 한두 가지 뚜렷한 감정에 집중하는 경향이 있음.

감정의 미묘한 경계나 애매함보다는, 분명한 감정 라벨을 선호하는 결과가 많음.

Gemini 모델  

감정의 복합성, 중립성, 애매함(‘그저 그런’, ‘만족스러운’, ‘불안한’, ‘우울한’ 등)을 더 섬세하게 포착함.

한 일기에 여러 감정이 혼재된 경우, 그 층위와 결까지 반영하려는 경향이 뚜렷함.

특히 ‘그저 그런’ 등 중립적 감정이나, ‘만족스러운’, ‘우울한’, ‘불안한’ 등 미묘한 감정의 구분이 잘 드러남.  

#### 2. 감정별 분포 및 일치도  
두 모델의 감정 분포는 상위 감정군에서는 겹치지만, ‘그저 그런’ 등 중립적 감정이나 ‘만족스러운’, ‘우울한’ 등 복합 감정에서 차이가 큼.

평균 Jaccard 유사도(0.526)는 두 모델이 절반 정도만 감정 조합에서 일치한다는 의미로, 감정 해석 방식이 상당히 다름을 보여줌.

Gemini는 감정의 애매함, 복합성, 무감정 상태를 더 자주 매핑하는 반면, Hugging Face는 보다 단순·직설적인 감정 분류에 집중함.

#### 3. 대표 사례 분석  
긍정 감정:  
Hugging Face는 ‘기쁜’ 등 즉각적 긍정에 집중, Gemini는 ‘만족스러운’, ‘감사한’ 등 여운과 복합적 긍정 감정을 더 자주 포착.

부정 감정:  
‘슬픈’과 ‘우울한’의 경계에서 Gemini는 맥락에 따라 더 세밀하게 구분, Hugging Face는 ‘슬픈’을 기본값처럼 넓게 적용.

중립/복합 감정:  
Gemini는 ‘그저 그런’ 등 중립적 감정을 적극적으로 매핑, Hugging Face는 거의 매핑하지 않음.

#### 4. 실제 일기 맥락과의 적합성  
 감정이 복합적이거나 애매한 일기에서는 Gemini의 결과가 더 현실적이고 섬세하게 느껴지는 경우가 많았음.  

 감정을 명확히 구분하거나, 빠르게 분류해야 하는 상황에서는 Hugging Face 결과가 더 단순·직관적일 수 있음.

#### 5. 객관적 결론  
감정의 복합성, 미묘함, 중립성까지 포착하고 싶다면 Gemini 모델이 더 적합함.

명확하고 단순한 감정 분류가 목적이라면 Hugging Face 모델이 유리함.

```python
from collections import Counter
import pandas as pd
import ast

# 안전하게 리스트로 변환
def safe_literal_eval(val):
    if isinstance(val, str):
        return ast.literal_eval(val)
    return val

df['exist_emotions'] = df['exist_emotions'].apply(safe_literal_eval)

# 평탄화 및 집계
all_emotions_df = [emotion for emotions in df['exist_emotions'] for emotion in emotions]
emotion_counts_df = Counter(all_emotions_df)
result_df = pd.DataFrame(emotion_counts_df.items(), columns=['emotion', 'count']).sort_values('count', ascending=False)

print("df 감정별 분포")
print(result_df)
```
